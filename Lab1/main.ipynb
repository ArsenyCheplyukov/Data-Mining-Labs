```python
def rozenbrock(x, y):
    return 100*(y - x**2)**2 + (1-x)**2

x_3d = np.linspace(-6, 6, 100)
y_3d = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x_3d, y_3d)
Z = rozenbrock(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('surface')
```

#### **Hist making**

```python
array_size = 1000
pockets_num = 15

rand_array = np.random.uniform(low=0.0, high=1.0, size=array_size)

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(rand_array, pockets_num, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.show()
```

#### **Normal destribution hist**

```python
normal_rand_array = np.random.normal(loc=10, scale=10, size=array_size) # loc-mean, scale-std

count, bins, ignored = plt.hist(normal_rand_array, density=True)
plt.show()
```

#### **Write into file**

```python
x_array = np.arange(0, 5, 0.1)
y_array = x_array**2 + x_array - 1

x_ = np.reshape(x_array, (x_array.size, -1))
y_ = np.reshape(y_array, (y_array.size, -1))
S = np.hstack((x_, y_)) # insert column, as vector to the end of matrix as column
with open('file.txt', 'w') as f:
    f.write(f"Function name: y = 2*x^2 + x - 1\n")
    for i in S:
        f.write(f"{i[0]} {i[1]}\n")
```

## **File read**

```python
current_x_array = np.array([])
current_y_array = np.array([])
with open('file.txt', 'r') as f:
    stuff = f.readlines()
    title_name = stuff[0]
    for row in stuff[1:]:
        a,b = row.split()
        current_x_array = np.append(current_x_array, float(a))
        current_y_array = np.append(current_y_array, float(b))
plt.grid()
plt.title(title_name)
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.plot(current_x_array, current_y_array, c="red")
N = current_y_array.size
plt.plot(current_x_array, current_x_array, c="blue")
plt.show()
```
